```python
import numpy as np
import matplotlib.pyplot as plt
from graphviz import Digraph
import scipy as sp
```

```python
### Visualize derivative 
fig = plt.figure(figsize = [3.2, 2.4])
ax = fig.add_subplot(111)

xs = np.arange(0, 5, .01)

def sq(x):
    y = .2*x**2 + 2
    return y

ax.plot(xs, sq(xs))

ax.set(xlim = [0, 5], ylim = [0, sq(5)], xticks = [1.5, 2.5], yticks = [sq(1.5), sq(2.5)], xlabel = 't', xticklabels = ['t', 't+$\delta$ t'], ylabel = 'I(t)', yticklabels = ['I(t)', 'I(t+$\delta$ t)'])

ax.hlines(y = sq(1.5), xmin = 0, xmax = 1.5, color = 'slategrey', linestyle = 'dashed')
ax.hlines(y = sq(2.5), xmin = 0, xmax = 2.5, color = 'slategrey', linestyle = 'dashed')

ax.vlines(x = 1.5, ymin = 0, ymax = sq(1.5), color = 'slategrey', linestyle = 'dashed')
ax.vlines(x = 2.5, ymin = 0, ymax = sq(2.5), color = 'slategrey', linestyle = 'dashed')

ax.fill_between(x = np.arange(1.5, 2.5, .01), y1 = sq(np.arange(1.5, 2.5, .01)), y2 = 0, color = 'salmon', alpha = .5)

fig.savefig('./figures/lecture1/derivative.png', dpi = 300)
```

```python
### SIR DAG

sir_dag = Digraph('dag', graph_attr={'rankdir': 'LR'})

for compartment in ['S', 'I', 'R']:
    sir_dag.node(compartment, compartment)

sir_dag.edge('S', 'I', 'λ')

sir_dag.edge('I', 'R', 'r')

sir_dag.render('./figures/lecture1/SIR_dag', format = 'png')
```

```python
### Plotting course of infection 

t = np.arange(-3, 3, .1)

i = np.exp(-t**2)

fig = plt.figure(figsize = [3.2, 2.4])

ax = fig.add_subplot(111)

ax.plot(t, i, c= 'salmon')
ax.fill_between(x = t, y1 = i, y2 = 0, color = 'salmon', alpha = .5)

ax.axvline(x = 0, ymin =0, ymax = 1/1.5 + .1, linestyle= 'dashed')

ax.set(xlim = [-4, 4], ylim = [0, 1.5], xticks = [], yticks = [])

ax.text(x = -2.5, y = .5, s = '$R_e > 1$')
ax.text(x = 1.25, y = .5, s = '$R_e < 1$')
ax.text(x = -.75, y = 1.2, s = '$R_e = 1$')

fig.savefig('./figures/lecture1/pandemic_timecourse.png', dpi = 300)
```

```python
### Visualizing k

k = 1
i_0 = 5

t = np.arange(0, 10, .2)

noise = sp.stats.norm(loc = 0, scale = 1).rvs(size = len(t), random_state = 111)

true = k * t + i_0
vals = true + noise

fig = plt.figure(figsize = [3.2, 2.4])
ax = fig.add_subplot(111)

ax.scatter(t[1:], vals[1:], c = 'salmon', alpha = .5, s= 10)
ax.plot(t, true)

ax.set(xlim = [0, 10], xticks = [], yticks = [i_0], yticklabels = ['$\ln(I_0)$'], xlabel = '$\ln(t)$', ylabel = '$\ln(I(t))$')

fig.savefig('./figures/lecture1/estimating_k.png', dpi = 300)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the linear function parameters ---
slope = 2
intercept = 5
num_points = 100

# --- 2. Generate a sequence of x-values ---
# We use np.linspace to get a range of evenly spaced numbers.
x = np.linspace(0, 10, num_points)

# --- 3. Calculate the "true" y-values from the linear function ---
# This is the perfect line without any noise.
y_true = slope * x + intercept

# --- 4. Generate random noise ---
# We use np.random.normal to create a set of random values
# with a specified mean and standard deviation.
# - loc=0: The noise is centered around zero (no bias).
# - scale=2: The standard deviation controls how spread out the noise is.
#            A larger value means more scatter around the line.
noise = np.random.normal(loc=0, scale=2, size=num_points)

# --- 5. Add the noise to the true y-values to get the noisy draws ---
y_noisy = y_true + noise

# --- 6. (Optional) Visualize the results ---
plt.figure(figsize=(8, 6))

# Plot the "true" linear function
plt.plot(x, y_true, color='red', linestyle='--', linewidth=2, label='True Linear Function')

# Plot the noisy data points as a scatter plot
plt.scatter(x, y_noisy, color='blue', alpha=0.6, label='Random Draws')

plt.title('Random Draws Around a Linear Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
